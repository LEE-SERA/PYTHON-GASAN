# 데이터프레임 인덱스 조작

    - set_index: 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
    - reset_index: 기존의 행 인덱스를 제고하고 인덱스를 데이터 열로 추가 

```python
import numpy as np
import pandas as pd
```

# set_index

```python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,#T를 넣어서 행열체인지 
                   columns=["C1", "C2", "C3", "C4"])
df1
```

```python
df2=df1.set_index('C1');df2 #C1열을 인덱스로 설정 
```

```python
df2.set_index('C2')
```

# reset_index
    인덱스를 보통의 자료열로 바꿈 

```python
df2.reset_index(drop=True)#drop을 설정하면 인덱스열을 보통의 자료열로 올리는 것이아니라 그냥버려버림 
```

# 다중인덱스 

```python
np.random.seed(0)
df3=pd.DataFrame(np.round(np.random.randn(5,4),2),
                 columns=[['A','A','B','B'],['C1','C2','C1','C2']])
df3
```

```python
#다중 인덱스에 이름 넣기
df3.columns.names=['Cidx1','Cidx2'];df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

# 행 인덱스와 열 인덱스 교환
    - stack(): 열 인덱스-> 행인덱스로 변환 
    - unstack():  행 인덱스 -> 열 인덱스로 변환 

```python
df4.stack('Cidx1')
```

```python
df4.stack(1)
```

```python
df4. unstack('Ridx2')
```

```python
df4.unstack(0)
```

# 다중 인덱스가 있는 경우의 인덱싱
    -데이터프레임이 다중 인덱스를 가지는 경우에는 인덱스가 ()로 둘러싸인 튜플이 되어야함


```python
df3
```

```python
#df.loc[행, 열 ]
df3.loc[0,('B','C1')]=100;df3 # 0행, B,C1 열 값을 100으로 한다.  
```

```python
#단, iloc 인덱서를 사용하는 경우에는 튜플 형태의 다중 인덱스를 사용할 수 없다. 
df3.iloc[0,2];df3
```

```python
df4
```

```python
df4.loc[:,('A','C')]
```

```python
df4.loc[('All','All'),:]=df4.sum();df4 #새로운 행을 추가해줌 
```

```python
#다중 인덱스의 인덱스 순서 교환 
#swaplevel(i,j,axis)
df5=df4.swaplevel('Ridx1','Ridx2');df5 #디폴트는 행 
```

```python
df6 = df4.swaplevel("Cidx1", "Cidx2", 1)
df6
```

# 다중 인덱스가 있는 경우의 정렬
    -데이터프레임을 sort_index로 정렬
    -level로 어떤 인덱스를 기준으로 정렬하는지설정  

```python
df5.sort_index(level=0)
```

```python
df6.sort_index(axis=1,level=0)
```
